### 주소를 입력하면 위도, 경도를 구해준다!

```python
import json 
import requests
from urllib.parse import urlparse

def address_to_lation(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?&query=' + address
    apikey = "3bded7c4b06d014ee5ebe60ed301e399"
    result = requests.get(urlparse(url).geturl(), headers={"Authorization" : "KakaoAK "+apikey})
    json_obj = result.json()
    result = json_obj['documents']
    #위도 경도 리턴해주기 
    if len(result) > 0:
        return result[0]['address']['x'], result[0]['address']['y']
    else:
        return None
    
```

```python
for i in range(0, len(gwangju_lights)):
    if gwangju_lights.iloc[i]['소재지지번주소'] != None:
        print(i, address_to_lation(gwangju_lights.iloc[i]['소재지지번주소']))
    else:
        print(i, address_to_lation(gwangju_lights.iloc[i]['소재지도로명주소']))
```
